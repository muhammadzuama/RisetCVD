```python
class DBSCANSMOTE(BaseOverSampler):
    """
    Menggunakan DBSCAN untuk memfilter data minoritas dari 'noise' (outlier).
    Setelah outlier dibuang, SMOTE diterapkan.
    """
    def __init__(self, eps=0.5, min_samples=5, random_state=None, **kwargs):
        super().__init__()
        self.eps = eps
        self.min_samples = min_samples
        self.random_state = random_state
        self.kwargs = kwargs

    def _fit_resample(self, X, y):
        target_stats = Counter(y)
        majority_class = max(target_stats, key=target_stats.get)

        X_clean, y_clean = [], []
        for c in target_stats.keys():
            mask = (y == c)
            Xc = X[mask]

            if c == majority_class or len(Xc) < self.min_samples:
                X_clean.append(Xc)
                y_clean.extend([c] * len(Xc))
            else:
                db = DBSCAN(eps=self.eps, min_samples=self.min_samples).fit(Xc)
                core_mask = db.labels_ != -1  # Membuang noise (label -1)

                if np.sum(core_mask) < 2:
                    core_mask = np.ones(len(Xc), dtype=bool)

                X_clean.append(Xc[core_mask])
                y_clean.extend([c] * np.sum(core_mask))

        if isinstance(X, pd.DataFrame):
            X_clean = pd.concat(X_clean, ignore_index=True)
        else:
            X_clean = np.vstack(X_clean)
        y_clean = np.array(y_clean)

        smote = SMOTE(random_state=self.random_state, **self.kwargs)
        return smote.fit_resample(X_clean, y_clean)
```

```python
class DBSCANROS(BaseOverSampler):
    """
    Menggunakan DBSCAN untuk menghapus outlier dari kelas minoritas,
    kemudian melakukan Random Over Sampler (ROS) pada titik data intinya.
    """
    def __init__(self, eps=0.5, min_samples=5, random_state=None):
        super().__init__()
        self.eps = eps
        self.min_samples = min_samples
        self.random_state = random_state

    def _fit_resample(self, X, y):
        target_stats = Counter(y)
        majority_count = max(target_stats.values())

        X_res, y_res = [], []

        for c, count in target_stats.items():
            mask = (y == c)
            Xc = X[mask]

            if count == majority_count:
                X_res.append(Xc)
                y_res.extend([c] * count)
            else:
                if len(Xc) < self.min_samples:
                    X_oversampled = resample(Xc, replace=True, n_samples=majority_count, random_state=self.random_state)
                    X_res.append(X_oversampled)
                    y_res.extend([c] * majority_count)
                    continue

                db = DBSCAN(eps=self.eps, min_samples=self.min_samples).fit(Xc)
                core_mask = db.labels_ != -1

                if np.sum(core_mask) == 0:
                    core_mask = np.ones(len(Xc), dtype=bool)

                X_core = Xc[core_mask]
                X_oversampled = resample(X_core, replace=True, n_samples=majority_count, random_state=self.random_state)
                X_res.append(X_oversampled)
                y_res.extend([c] * majority_count)

        if isinstance(X, pd.DataFrame):
            X_res = pd.concat(X_res, ignore_index=True)
        else:
            X_res = np.vstack(X_res)
        return X_res, np.array(y_res)
```

```python
class KMeansROS(BaseOverSampler):
    """
    Mengelompokkan data minoritas menjadi beberapa cluster menggunakan KMeans,
    lalu melakukan oversampling (ROS) secara seimbang pada setiap cluster.
    """
    def __init__(self, n_clusters=5, random_state=None):
        super().__init__()
        self.n_clusters = n_clusters
        self.random_state = random_state

    def _fit_resample(self, X, y):
        target_stats = Counter(y)
        majority_count = max(target_stats.values())

        X_res, y_res = [], []

        for c, count in target_stats.items():
            mask = (y == c)
            Xc = X[mask]

            if count == majority_count:
                X_res.append(Xc)
                y_res.extend([c] * count)
            else:
                n_clusters = min(self.n_clusters, len(Xc))
                if n_clusters > 1:
                    kmeans = KMeans(n_clusters=n_clusters, random_state=self.random_state, n_init=10).fit(Xc)
                    labels = kmeans.labels_
                else:
                    labels = np.zeros(len(Xc))

                samples_per_cluster = majority_count // n_clusters
                remainder = majority_count % n_clusters

                for i in range(n_clusters):
                    cluster_mask = (labels == i)
                    X_cluster = Xc[cluster_mask]
                    if len(X_cluster) == 0: continue

                    n_samples = samples_per_cluster + (1 if i < remainder else 0)
                    X_oversampled = resample(X_cluster, replace=True, n_samples=n_samples, random_state=self.random_state)
                    X_res.append(X_oversampled)
                    y_res.extend([c] * n_samples)

        if isinstance(X, pd.DataFrame):
            X_res = pd.concat(X_res, ignore_index=True)
        else:
            X_res = np.vstack(X_res)
        return X_res, np.array(y_res)
```

```python
# ============================================================
# LANGKAH 5: BPSO Feature Selection dengan XGBoost sebagai estimator
# ============================================================
import time
import numpy as np
from pyswarms.discrete import BinaryPSO
from sklearn.model_selection import cross_val_score
# --- Scaling (opsional tapi direkomendasikan) ---
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

N_CLASSES  = len(np.unique(y_train))
CV_FOLDS   = 3   # kurangi jika data besar supaya tidak terlalu lama

# ── Fungsi Objektif ──────────────────────────────────────────
def pso_objective_xgb(mask):
    """
    mask  : (n_particles, n_features) — nilai biner 0/1
    return: (n_particles,) — cost = 1 - accuracy (minimasi)
    """
    scores = []
    for i in range(mask.shape[0]):
        selected = np.where(mask[i] > 0.5)[0]

        # Jika tidak ada fitur terpilih → penalti maksimal
        if len(selected) == 0:
            scores.append(1.0)
            continue

        X_sel = X_train_scaled[:, selected]

        xgb = XGBClassifier(
            objective      = "multi:softprob",
            num_class      = N_CLASSES,
            n_estimators   = 100,          # lebih sedikit supaya cepat di PSO
            learning_rate  = 0.1,
            max_depth      = 4,
            subsample      = 0.8,
            colsample_bytree = 0.8,
            eval_metric    = "mlogloss",
            use_label_encoder = False,
            random_state   = 42,
            verbosity      = 0,
            n_jobs         = -1,
        )

        acc = cross_val_score(
            xgb,
            X_sel, y_train,
            cv      = CV_FOLDS,
            scoring = "accuracy",
            n_jobs  = -1,
        ).mean()

        scores.append(1.0 - acc)

    return np.array(scores)


# ── Jalankan BinaryPSO ───────────────────────────────────────
n_features = X_train_scaled.shape[1]

options = {
    'c1': 2.0,   # cognitive coefficient
    'c2': 2.0,   # social coefficient
    'w' : 0.9,   # inertia weight
    'k' : 10,    # jumlah tetangga (ring topology)
    'p' : 2,     # p-norm (2 = euclidean)
}

optimizer = BinaryPSO(
    n_particles = 30,
    dimensions  = n_features,
    options     = options,
)

print("\n" + "=" * 55)
print("BPSO FEATURE SELECTION (XGBoost Estimator)")
print("=" * 55)
print(f"Jumlah fitur     : {n_features}")
print(f"Jumlah partikel  : 30")
print(f"Iterasi          : 50")
print(f"CV folds         : {CV_FOLDS}")
print("Mulai optimisasi ...\n")

t0 = time.time()
# best_cost, best_mask = optimizer.optimize(pso_objective_xgb, iters=10) # bagian iterasi al
best_cost, best_mask = optimizer.optimize(pso_objective_xgb, iters=50) # bagian iterasi al
elapsed = time.time() - t0

selected_idx = np.where(best_mask > 0.5)[0]

# Terapkan ke data
X_train_pso = X_train_scaled[:, selected_idx]
X_test_pso  = X_test_scaled[:,  selected_idx]

# Nama fitur terpilih (jika X_train adalah DataFrame)
if hasattr(X_train, 'columns'):
    selected_features = X_train.columns[selected_idx].tolist()
    print(f"Fitur terpilih   : {selected_features}")
else:
    print(f"Indeks terpilih  : {selected_idx.tolist()}")

print(f"\nJumlah fitur terpilih : {len(selected_idx)} / {n_features}")
print(f"Best cost (1-acc)     : {best_cost:.4f}")
print(f"Estimated accuracy    : {(1 - best_cost) * 100:.2f}%")
print(f"Waktu PSO             : {elapsed:.1f} detik")
print(f"Shape X_train_pso     : {X_train_pso.shape}")
print(f"Shape X_test_pso      : {X_test_pso.shape}")
```

```python
from imblearn.combine import SMOTEENN
from collections import Counter

# Target names sesuai label CVD_SEVERITY
TARGET_NAMES = ['0', '1', '2', '3']

# ── Helper: evaluate satu kombinasi sampler+model ────────────
def evaluate_one_pso(models, X_tr, y_tr, X_te, y_te, label, target_names=TARGET_NAMES):
    rows = []
    for name, model in models.items():
        print(f"   🔧 {name} | {label}")
        model.fit(X_tr, y_tr)
        y_pred  = model.predict(X_te)
        y_proba = model.predict_proba(X_te)

        acc = accuracy_score(y_te, y_pred)
        f1  = f1_score(y_te, y_pred, average='macro')
        auc = roc_auc_score(y_te, y_proba, multi_class='ovr', average='macro')
        cm  = confusion_matrix(y_te, y_pred)

        print(f"      ✅ Acc={acc:.4f}  F1={f1:.4f}  AUC={auc:.4f}")
        print(f"      CM:\n{cm}")
        print(classification_report(y_te, y_pred, target_names=target_names))

        rows.append({
            "Model"   : name,
            "Sampler" : label,
            "Accuracy": round(acc, 4),
            "F1-Score": round(f1,  4),
            "ROC-AUC" : round(auc, 4),
        })
    return pd.DataFrame(rows)


# ============================================================
# EKSPERIMEN 1 — RAW (tanpa resampling, fitur PSO)
# ============================================================
print("\n" + "=" * 60)
print("⚙️  SAMPLER: RAW | Fitur: PSO")
print("=" * 60)
print(f"Distribusi y_train : {sorted(Counter(y_train).items())}")

df_raw_pso = evaluate_one_pso(
    MODELS,
    X_train_pso, y_train,
    X_test_pso,  y_test,
    label="raw_pso"
)

df_raw_pso.to_csv("hasil_evaluasi_raw_pso.csv", index=False)
print("📁 Disimpan → hasil_evaluasi_raw_pso.csv")


# ============================================================
# EKSPERIMEN 2 — SMOTEENN (fitur PSO)
# ============================================================
print("\n" + "=" * 60)
print("⚙️  SAMPLER: SMOTEENN | Fitur: PSO")
print("=" * 60)

smoteenn = SMOTEENN(random_state=42)
X_train_smoteenn, y_train_smoteenn = smoteenn.fit_resample(X_train_pso, y_train)

print(f"Distribusi sebelum : {sorted(Counter(y_train).items())}")
print(f"Distribusi sesudah : {sorted(Counter(y_train_smoteenn).items())}")
print(f"Shape X_train      : {X_train_smoteenn.shape}")

df_smoteenn_pso = evaluate_one_pso(
    MODELS,
    X_train_smoteenn, y_train_smoteenn,
    X_test_pso,       y_test,
    label="smoteenn_pso"
)

df_smoteenn_pso.to_csv("hasil_evaluasi_smoteenn_pso.csv", index=False)
print("📁 Disimpan → hasil_evaluasi_smoteenn_pso.csv")


# ============================================================
# GABUNG DAN TAMPILKAN SEMUA HASIL
# ============================================================
df_all_pso = pd.concat([df_raw_pso, df_smoteenn_pso], ignore_index=True)
df_all_pso.to_csv("hasil_evaluasi_ALL_pso.csv", index=False)

print("\n" + "=" * 60)
print("📊 RINGKASAN HASIL (diurutkan F1-Score)")
print("=" * 60)
print(df_all_pso.sort_values("F1-Score", ascending=False).to_string(index=False))
```

```python
def apply_feature_selection(X_train, X_test, y_train, top_k=TOP_K):
    print(f"\n📌 Feature Selection: memilih {top_k} fitur terbaik dengan XGBoost...")
    xgb_fs = XGBClassifier(n_estimators=100, random_state=42,
                            eval_metric='mlogloss', verbosity=0)
    xgb_fs.fit(X_train, y_train)

    feat_df = pd.DataFrame({
        'feature'   : X_train.columns if hasattr(X_train, 'columns') else [f"f{i}" for i in range(X_train.shape[1])],
        'importance': xgb_fs.feature_importances_,
    }).sort_values('importance', ascending=False)

    selected = feat_df.head(top_k)['feature'].tolist()
    print(f"   Top fitur: {selected[:5]} ... ({top_k} total)")

    if hasattr(X_train, 'columns'):
        return X_train[selected], X_test[selected]
    else:
        idx = feat_df.head(top_k).index.tolist()
        return np.array(X_train)[:, idx], np.array(X_test)[:, idx]

```

```python
def evaluate_one(models, X_tr, y_tr, X_te, y_te, label, target_names=TARGET_NAMES):
    rows = []
    for name, model in models.items():
        print(f"   🔧 {name} | {label}")
        model.fit(X_tr, y_tr)
        y_pred  = model.predict(X_te)
        y_proba = model.predict_proba(X_te)

        acc = accuracy_score(y_te, y_pred)
        f1  = f1_score(y_te, y_pred, average='macro')
        auc = roc_auc_score(y_te, y_proba, multi_class='ovr', average='macro')
        cm  = confusion_matrix(y_te, y_pred)

        print(f"      ✅ Acc={acc:.4f}  F1={f1:.4f}  AUC={auc:.4f}")
        print(f"      CM:\n{cm}")
        print(classification_report(y_te, y_pred, target_names=target_names))

        rows.append({
            "Model"   : name,
            "Dataset" : label,
            "Accuracy": round(acc, 4),
            "F1-Score": round(f1,  4),
            "ROC-AUC" : round(auc, 4),
        })
    return pd.DataFrame(rows)
```

```python
def run_pipeline(X_train, X_test, y_train, y_test,
                 models=MODELS, samplers=SAMPLERS,
                 use_fs=USE_FS, top_k=TOP_K):

    print("=" * 60)
    print("🚀 MEMULAI PIPELINE")
    print("=" * 60)

    # Feature selection
    if use_fs:
        X_tr_base, X_te_base = apply_feature_selection(X_train, X_test, y_train, top_k)
        fs_tag = "fs"
    else:
        X_tr_base, X_te_base = X_train, X_test
        fs_tag = "nonfs"

    all_results = []

    for sampler_name, sampler in samplers.items():
        label = f"{sampler_name}_{fs_tag}"
        print(f"\n{'─'*50}")
        print(f"⚙️  Sampler: {sampler_name.upper()} | FS: {use_fs}")

        # Balancing
        if sampler is None:
            X_tr, y_tr = X_tr_base, y_train
        else:
            X_tr, y_tr = sampler.fit_resample(X_tr_base, y_train)
            dist = sorted(Counter(y_tr).items())
            print(f"   Distribusi setelah resample: {dist}")

        # Train & evaluate
        df = evaluate_one(models, X_tr, y_tr, X_te_base, y_test, label)
        all_results.append(df)

        # Simpan CSV per sampler
        fname = f"hasil_evaluasi_{label}.csv"
        df.to_csv(fname, index=False)
        print(f"   📁 Disimpan → {fname}")

    # Gabung semua hasil
    final_df = pd.concat(all_results, ignore_index=True)
    final_df.to_csv("hasil_evaluasi_SEMUA.csv", index=False)

    print("\n" + "=" * 60)
    print("✅ PIPELINE SELESAI")
    print("=" * 60)
    print(final_df.sort_values("F1-Score", ascending=False).to_string(index=False))

    return final_df
```
