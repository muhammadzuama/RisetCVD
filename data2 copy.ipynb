```python
# Cek class apa saja yang ada di data
print("Classes di y_train:", sorted(y_train.unique()))
print("Classes di y_test :", sorted(y_test.unique()))
print("Jumlah class       :", y_test.nunique())
```

```python
from sklearn.metrics import (
    accuracy_score, classification_report,
    roc_auc_score, confusion_matrix,f1_score
)
 
# ── Ensemble Models ─────────────────────────────────────────────────────────
from sklearn.ensemble import (
    RandomForestClassifier,
    GradientBoostingClassifier,
    ExtraTreesClassifier,
)

from lightgbm import LGBMClassifier
```

```python
n_classes = 4
```

```python
models = {
 
    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        max_features="sqrt",
        class_weight="balanced",   # bagus jika kelas tidak seimbang
        n_jobs=-1,
        random_state=42,
    ),
 
    "XGBoost": XGBClassifier(
        objective="multi:softprob",   # ← wajib untuk multi-class
        num_class=n_classes,          # ← jumlah kelas
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        use_label_encoder=False,
        eval_metric="mlogloss",       # ← multi-class log loss
        random_state=42,
        verbosity=0,
    ),
 
    # "GBM (Gradient Boosting)": GradientBoostingClassifier(
    #     # scikit-learn GBM otomatis pakai One-vs-Rest untuk multi-class
    #     n_estimators=300,
    #     learning_rate=0.05,
    #     max_depth=4,
    #     subsample=0.8,
    #     min_samples_split=2,
    #     min_samples_leaf=1,
    #     random_state=42,
    # ),
 
    "LightGBM": LGBMClassifier(
        objective="multiclass",       # ← wajib untuk multi-class
        num_class=n_classes,          # ← jumlah kelas
        n_estimators=300,
        learning_rate=0.05,
        max_depth=-1,
        num_leaves=31,
        subsample=0.8,
        colsample_bytree=0.8,
        class_weight="balanced",
        n_jobs=-1,
        random_state=42,
        verbose=-1,
    ),
 
    "Extra Trees": ExtraTreesClassifier(
        n_estimators=300,
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1,
        max_features="sqrt",
        class_weight="balanced",
        n_jobs=-1,
        random_state=42,
    ),
}
 
```

```python
# ============================================================
# FUNCTION: Train & Evaluate
# ============================================================
def train_evaluate(
    models: dict,
    X_train, y_train,
    X_test,  y_test,
    dataset_label: str = "default",     # nama dataset, misal "smotenn", "ctgan", "raw"
    export_csv: bool   = True,
    target_names: list = ['0','1','2','3']
):
    results     = {}
    summary_rows = []

    for name, model in models.items():
        print(f"\n🔧 Training: {name} | Dataset: {dataset_label} ...")

        # ── Fit ──────────────────────────────────────────────
        model.fit(X_train, y_train)

        # ── Predict ──────────────────────────────────────────
        y_pred  = model.predict(X_test)
        y_proba = model.predict_proba(X_test)

        # ── Metrics ──────────────────────────────────────────
        acc     = accuracy_score(y_test, y_pred)
        f1      = f1_score(y_test, y_pred, average='macro')
        auc     = roc_auc_score(y_test, y_proba, multi_class='ovr', average='macro')
        cm      = confusion_matrix(y_test, y_pred)
        report  = classification_report(y_test, y_pred, target_names=target_names)

        # ── Simpan detail ─────────────────────────────────────
        results[name] = {
            "model"          : model,
            "Accuracy"       : acc,
            "F1-Score"       : f1,
            "ROC-AUC"        : auc,
            "confusion_matrix": cm,
            "report"         : report,
            "dataset"        : dataset_label,
        }

        summary_rows.append({
            "Model"        : name,
            "Dataset"      : dataset_label,
            "Accuracy"     : round(acc, 4),
            "F1-Score"     : round(f1,  4),
            "ROC-AUC"      : round(auc, 4),
        })

        # ── Print ─────────────────────────────────────────────
        print(f"   ✅ Accuracy: {acc:.4f}  |  F1-Score: {f1:.4f}  |  ROC-AUC: {auc:.4f}")
        print(f"\nConfusion Matrix:\n{cm}")
        print(f"\nClassification Report:\n{report}")

    # ── Buat DataFrame ringkasan ──────────────────────────────
    results_df = pd.DataFrame(summary_rows)

    # ── Export CSV ────────────────────────────────────────────
    if export_csv:
        filename = f"hasil_evaluasi_{dataset_label}.csv"
        results_df.to_csv(filename, index=False)
        print(f"\n📁 Hasil disimpan → {filename}")

    return results_df, results
```

```python
from imblearn.over_sampling import SMOTE, ADASYN
# ── SMOTE FS ─────────────────────────────────────────────────────────────────────
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_fs, y_train)
print(f"SMOTE → {sorted(dict(zip(*np.unique(y_train_smote, return_counts=True))).items())}")

# ── ADASYN FS ────────────────────────────────────────────────────────────────────
adasyn = ADASYN(random_state=42)
X_train_adasyn, y_train_adasyn = adasyn.fit_resample(X_train_fs, y_train)
print(f"ADASYN → {sorted(dict(zip(*np.unique(y_train_adasyn, return_counts=True))).items())}")


# ── SMOTE Non FS─────────────────────────────────────────────────────────────────────
X_train_smote_nonFs, y_train_smote_nonFs = smote.fit_resample(X_train, y_train)
print(f"SMOTE → {sorted(dict(zip(*np.unique(y_train_smote_nonFs, return_counts=True))).items())}")

# ── ADASYN Non FS ────────────────────────────────────────────────────────────────────
X_train_adasyn_nonFs, y_train_adasyn_nonFs = adasyn.fit_resample(X_train, y_train)
print(f"ADASYN → {sorted(dict(zip(*np.unique(y_train_adasyn_nonFs, return_counts=True))).items())}")
```
